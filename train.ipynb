```python
from monai.networks.nets import UNet
from monai.networks.layers import Norm
from monai.losses import DiceLoss, DiceCELoss

import torch
from preprocess import prepare
from utilities import train
```

```python
data_dir = 'E:/Hitesh/Univ/self_learning/ML_projects/Medical_seg_data/Train_test_data'
model_dir = "E:/Hitesh/Univ/self_learning/ML_projects/Medical_seg_data/model_results"
data_in = prepare(data_dir, cache=True)

device = torch.device("cuda:0")
model = UNet(
    spatial_dims=3,
    in_channels=1,
    out_channels=2,
    channels=(16, 32, 64, 128, 256),
    strides=(2, 2, 2, 2),
    num_res_units=2,
    norm=Norm.BATCH,
).to(device)
```

```python
#loss_function = DiceCELoss(to_onehot_y=True, sigmoid=True, squared_pred=True, ce_weight=calculate_weights(1792651250,2510860).to(device))
loss_function = DiceLoss(to_onehot_y=True, sigmoid=True, squared_pred=True)
optimizer = torch.optim.Adam(model.parameters(), 1e-5, weight_decay=1e-5, amsgrad=True)
```

```python
if __name__ == '__main__':
    train(model, data_in, loss_function, optimizer, 200, model_dir)
```
