```python
import os

# Function to count the total number of files in a folder
def count_files_in_folder(folder_path):
    if not os.path.exists(folder_path):
        print(f"The folder '{folder_path}' does not exist.")
        return

    file_count = 0

    for root, _, files in os.walk(folder_path):
        file_count += len(files)
        print(f"Folder '{root}' contains {len(files)} files.")

    print(f"Total number of files in '{folder_path}': {file_count}")

# Provide the path to the folder you want to analyze
folder_path = r"E:\Hitesh\Univ\self_learning\ML_projects\Medical_seg_data\dicom\images"

count_files_in_folder(folder_path)
```

```python
in_path = r"E:\Hitesh\Univ\self_learning\ML_projects\Medical_seg_data\dicom\images"
out_path = r"E:\Hitesh\Univ\self_learning\ML_projects\Medical_seg_data\dicom\dicom_groups\images"
```

```python
from glob import glob
import shutil
import os
```

```python
for pat in glob(in_path+'/*'):
    pat_name = os.path.basename(os.path.normpath(pat))
    num_fld = int(len(glob(pat+'/*'))/64)

    for i in range(num_fld):
        op_name = os.path.join(out_path, pat_name+'_'+str(i))
        os.mkdir(op_name)
        for i, file in enumerate(glob(pat+'/*')):
            if i==64 + 1:
                break
            shutil.move(file, op_name)
```
